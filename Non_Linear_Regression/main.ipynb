```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("ice_cream_nonlinear.csv")
```

```python
x=df[['Temperature','IceCream_Sales']]
y=df['Profit']
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=43)
model=LinearRegression()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=1)
updated_x_train=poly.fit_transform(x_train)
x_train_poly=poly.transform(x_test)

model.fit(updated_x_train, y_train)
```

```python
updated_x_train[:,1]
```

```python
ypred_train=model.predict(x_train_poly)
ypred_train
```

```python
plt.scatter(updated_x_train[:,1], y_train)
plt.scatter(updated_x_train[:,1],ypred_train)
```
